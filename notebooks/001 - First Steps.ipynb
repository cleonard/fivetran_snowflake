```python
import datetime
import os
from pathlib import Path
from pprint import pp

from dotenv import load_dotenv
import requests

load_dotenv()

BASE_URL = "https://newsapi.org/v2/everything"
TOPIC = "Department of Education"

headers = {"Authorization": f"Bearer {os.environ['NEWSAPI_KEY']}", "accept": "application/json"}

from_ts = datetime.datetime.now() - datetime.timedelta(days=1)
now = datetime.datetime.now()
to_ts = now.strftime("%Y-%m-%dT%H:%M:%S")
params = {
    "from": from_ts,
    "to": to_ts,
    "page": "1",
    "language": "en",
    "sortBy": "publishedAt",
    "pageSize": "100",
}
params["q"] = TOPIC

r = requests.get(BASE_URL, headers=headers, params=params)
r
```
